Para investigar os tipos disponíveis, podemos usar

```julia
function showtypetree(T, level=0)
    println("\t" ^ level, T)
    for t in subtypes(T)
        if t != Any
            showtypetree(t, level+1)
        end
   end
end

showtypetree(Any)
```

## Vetores

Ao criar um vetor, ele nunca está vazio:

```julia
Array{Int64}(3)
```

```julia
lista = Array{String}(3)
```

Para acessar elementos da lista, começamos com índice 1 (mas pode-se mudar arbitrariamente os índices da lista)

```julia
lista[1] = "teste"
```

Para acrescentar um elemento ao final, podemos usar **push!** (o *bang* - ponto de exclamação - informa que a função está modificando sua entrada)

```julia
push!(lista, "teste")
```

Como no Python, podemos acessar os elementos da string como uma lista:

```julia
lista[1][2]
```

Mas na segunda entrada da lista, não temos uma string, temos o valor **#undef**

```julia
lista[2][1]
```

Para saber mais sobre qualquer função, use **?**

```julia
? push
```

Para acrescentar elementos no início da lista, use

```julia
unshift!(lista, "ops")
```

### Iteradores e comprehensions

Podemos gerar listas de valores igualmente espaçados (oi, MATLAB!)

```julia
linspace(0, 10, 11)
```

```julia
typeof(linspace(0,10,11))
```

Como o **linspace** é um iterador, para acessarmos os valores gerados usamos **collect**:

```julia
collect(linspace(0,10,11))
```

Podemos usar espaçamento arbitrário:

```julia
vetor = collect(range(0, 2, 10))
```

Podemos fazer operações elemento-a-elemento em um vetor usando o .

```julia
vetor .> 5
```

Também temos comprehensions:

```julia
[n^2 for n in 1:3]
```

E coisas interessantes: 

```julia
[a * b for a in 1:3, b in 1:3]
```

Se tentarmos fazer isso em Python,

[a*b for a in range(0,4) for b in range(0,4)]

retorna uma lista de 9 elementos.

### Álgebra Linear

Temos todas as facilidades para lidar com matrizes e álgebra linear (herdadas da Lapack e da BLAS):

```julia
A = [1 2; 3 4]; b = [5, 6];
```

(observe que ponto-e-vírgula no final da linha impede o display do comando - oi, MATLAB!)

Suponha que queremos resolver o sistema Ax=b:

```julia
x = A\b
```

Da documentação:

**Base.LinAlg.factorize**: factorize(A)

Compute a convenient factorization of A, based upon the type of the input matrix. if A is passed as a generic matrix. factorize checks every element of A to verify/rule out each property. It will short-circuit as soon as it can rule out symmetry/triangular structure. 

| Properties of A	  | type of factorization |
|:--------------------|:--------------------- |
| Positive-definite   | Cholesky (see cholfact) |
| Dense Symmetric/Hermitian | Bunch-Kaufman (see bkfact) |
| Sparse Symmetric/Hermitian | LDLt (see ldltfact) |
| Triangular | Triangular |
|Diagonal | Diagonal|
|Bidiagonal | Bidiagonal|
|Tridiagonal | LU (see lufact)|
|Symmetric real tridiagonal | LDLt (see ldltfact)|
|General square | LU (see lufact)|
|General non-square | QR (see qrfact)|

If factorize is called on a Hermitian positive-definite matrix, for instance, then factorize will return a Cholesky factorization.

## Macros

Os macros começam sempre por @ e em geral representam expressões que serão avaliadas/parseadas/sofrerão alguma alteração no momento da execução.

```julia
@time for i = 1:10000000 x = x+i end
```

Também podemos definir macros mais complexos usando o ambiente macro:

```julia
macro sayhello(name)
    return :(println("Hello, ", $name) )
end
```

```julia
@sayhello("there")
```

Para verificarmos o que será executado por um macro, podemos usar

```julia
macroexpand(:(@sayhello("there")))
```

Também é possível usar *expressions* que podem ser parseadas no momento da execução (não vamos tratar disso!)
